```python
# Multi level inheritance

class fruitsmarket:
  def hometown(self):
    print("fruitsmarket class and hometown")

class parent(fruitsmarket):
  def appale(self):
    print("parent class with appale")

class child(parent):
  def orange(self):
    print("child class with orange")

class grandchild(child):
  def grapes(self):
    print("this is grandchild class with grapes")

a1 = grandchild()
a1.appale()

a1.orange()
a1.grapes()
```

```python
# multiple inheritance

class father:
  def homeking(self):
    print("homeking function of  father class")

class mother:
  def homemqueen(self):
    print("homequeen function of mother class")

class child(father,mother):
  def own(self):
    print("this is child class in multiple")

a1 =child()

a1.homeking()
a1.homemqueen()
a1.own()
```

```python
# Hierachical inheritance

class father:
  def homeking(self):
    print("homeking function of  father class")

class child1(father):
  def quality(self):
    print("child quality")

class child2 (father):
  def dailytask(self):
    print("child2 dailytask")

a1 = child1()
a2 = child2()
a1.homeking()






```

```python
#Hybrid inheritance

class father:
  def homeking(self):
    print("homeking function of  father class")

class mother:
  def homemqueen(self):
    print("homequeen function of mother class")

class child1(father, mother):
  def quality(self):
    print("child quality")

class child2 (child1):
  def dailytask(self):
    print("child2 dailytask")




a1.homeking()
a1.quality()
a1 = child1()
a2 = child2()

```

```python
# Multi level inheritance

class asthma:
  def copd(self):
    print("asthma class of copd")

class diagnosis(asthma):
  def salbutamol(self):
    print("salbutamol drug use for asthma ")

class symptom(diagnosis):
  def oxygen(self):
    print("Oxygen use for asthma")

class sing(symptom):
  def treatment(self):
    print("This is all sign and symptom with tratment")

a1 = sing()
a1.salbutamol()

a1.oxygen()
a1.treatment()
```
